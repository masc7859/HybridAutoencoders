```python
import torch
import torch.optim

import CIFAR_unet as C

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
device
```

```python
model = C.CIFARAutoEncoderUNet(20).to(device)
train_load, test_load = C.get_cifar_loaders(download=False)
epoch = 0
opt = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
for _ in range(50):
    C.train(model, device, train_load, opt, epoch)
    C.test(model, device, train_load)
    epoch += 1
```
