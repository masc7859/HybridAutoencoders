```python
import torch
import torch.optim

#import CIFAR_v2 as C
import CIFAR_v3 as C

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
model = C.CIFARAutoEncoderV3(80).to(device)
#model = C.CIFARAutoEncoderV2(512).to(device)
epoch_enc = 0
epoch_dec = 0
train_load, test_load = C.get_cifar_loaders(download=False)
opt_enc = torch.optim.Adam(model.classifier.parameters(), lr=0.001)
opt_dec = torch.optim.Adam(model.decoder.parameters(), lr=0.001)
```

```python
for _ in range(15):
    epoch_enc += 1
    C.train_encoder(model, device, train_load, opt_enc, epoch_enc)
    C.test_encoder(model, device, test_load)
```

```python
for _ in range(50):
    epoch_dec += 1
    C.train_decoder(model, device, train_load, opt_dec, epoch_dec)
    C.test_decoder(model, device, test_load)
```
