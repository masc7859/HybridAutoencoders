```python
import torch
import torch.optim

import MNIST
import CIFAR_v1
import CIFAR_v2

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
mnist_model = MNIST.MnistAutoEncoder(255).to(device)
mnist_epoch = 0
mnist_train_loader, mnist_test_loader = MNIST.get_mnist_loaders()
mnist_opt = torch.optim.SGD(mnist_model.parameters(), lr=0.01, momentum=0.5)
```

```python
for _ in range(10):
    mnist_epoch += 1
    MNIST.train(mnist_model, device, mnist_train_loader, mnist_opt, mnist_epoch)
    MNIST.test(mnist_model, device, mnist_test_loader)
```

```python
MNIST.show_true_and_recreated_imgs(mnist_model, mnist_train_loader, device)
```

```python
MNIST.show_transition(mnist_model, mnist_test_loader, device)
```

```python
cifar_v1_model = CIFAR_v1.CIFARAutoEncoderV1(255).to(device)
cifar_v1_epoch = 0
cifar_v1_train_loader, cifar_v1_test_loader = CIFAR_v1.get_cifar_loaders(download=False)
cifar_v1_opt = torch.optim.SGD(cifar_v1_model.parameters(), lr=0.01, momentum=0.5)
```

```python
for _ in range(10):
    cifar_v1_epoch += 1
    CIFAR_v1.train(cifar_v1_model, device, cifar_v1_train_loader, cifar_v1_opt, cifar_v1_epoch)
    CIFAR_v1.test(cifar_v1_model, device, cifar_v1_test_loader)
```

```python
CIFAR_v1.show_true_and_recreated_imgs(cifar_v1_model, cifar_v1_test_loader, device)
```

```python
cifar_v2_model = CIFAR_v2.CIFARAutoEncoderV2(512).to(device)
cifar_v2_encoder_epoch = 0
cifar_v2_decoder_epoch = 0
cifar_v2_train_loader, cifar_v2_test_loader = CIFAR_v2.get_cifar_loaders(download=False)
cifar_v2_encoder_opt = torch.optim.Adam(cifar_v2_model.classifier.parameters(), lr=0.001)
cifar_v2_decoder_opt = torch.optim.Adam(cifar_v2_model.decoder.parameters(), lr=0.001)
```

```python
for _ in range(10):
    cifar_v2_encoder_epoch += 1
    CIFAR_v2.train_encoder(cifar_v2_model, device, cifar_v2_train_loader,
                           cifar_v2_encoder_opt, cifar_v2_encoder_epoch)
    CIFAR_v2.test_encoder(cifar_v2_model, device, cifar_v2_test_loader)
```

```python
for _ in range(10):
    cifar_v2_decoder_epoch += 1
    CIFAR_v2.train_decoder(cifar_v2_model, device, cifar_v2_train_loader,
                           cifar_v2_encoder_opt, cifar_v2_encoder_epoch)
    CIFAR_v2.test_decoder(cifar_v2_model, device, cifar_v2_test_loader)
```

```python
CIFAR_v2.show_true_and_recreated_imgs(cifar_v2_model, cifar_v2_train_loader, device)
```

```python
CIFAR_v2.show_transition(cifar_v2_model, cifar_v2_train_loader, device)
```
